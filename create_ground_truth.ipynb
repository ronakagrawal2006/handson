```python
marked_data_v1=pd.read_csv('sample_data_v1.csv')
```

```python
marked_data_v1.columns[~marked_data_v1.columns.isin(marked_data_v1.filter(like='Unnamed').columns)]
```

```python
marked_data_v2=marked_data_v1[marked_data_v1.columns[~marked_data_v1.columns.isin(marked_data_v1.filter(like='Unnamed').columns)]]
marked_data_v3=marked_data_v2[~marked_data_v2['_id'].isna()]
```

```python
import numpy as np
marked_data_v4=marked_data_v3[['_id','C1','C2']].replace('None',np.nan)
```

```python
marked_data_v4.head(2)
```

```python
def remove_nan(mylist):
    return [str(x) for x in mylist if x==x]

# remove_nan(['Sensetime', np.nan])
```

```python
    
marked_data_v4.loc[:,'companies']=marked_data_v4[['C1','C2']].apply(lambda x: x.values.tolist(),axis=1)
marked_data_v4['companies']=marked_data_v4.apply(lambda x: remove_nan(x['companies']),axis=1)
```

```python
marked_data_v4.drop(['C1','C2'],axis=1,inplace=True)
```

```python
marked_data_vn=marked_data_v4
```

```python
marked_data_vn.to_csv('marked_data_vn.csv',index=False)
```
