```python
import spacy
nlp = spacy.load('en_core_web_sm')
def get_orgs(x):
    doc = nlp(x) 
    ents=[]
    if doc.ents: 
        for ent in doc.ents: 
            if ent.label_=='ORG':
                ents.append(ent.text) 
    else: 
        pass
#         print('No named entities found.')
    return ents
```

```python
def get_orgs_list(y):
    t=[get_orgs(x) for x in y]
    flat_list = [item for sublist in t for item in sublist]
    return flat_list
```

```python
org_names_raw_v1=content_series_vn['sentences'].apply(lambda x: get_orgs_list(x)).to_frame(name='companies_list')
```
