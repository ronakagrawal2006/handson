```python
news_data_raw_v1=pd.read_csv('news_data.csv')
news_data_raw_v1.columns

#drop not required columns here
news_data_raw_v1.drop(['source','theme','published_on'],axis=1,inplace=True)# 
news_data_raw_v1
```

```python
content_series=news_data_raw_v1['description'].to_frame()
```

```python
# remove punctuations
# lower, stemmer, lemmatizer
```

```python

from nltk import sent_tokenize
from nltk.tokenize import word_tokenize
# sentences = sent_tokenize(text)
# content_series.loc[:,'sentences'] = content_series['description'].apply(lambda x: sent_tokenize(x))
# content_series.iloc[2]['sentences']
content_series.loc[:,'words'] = content_series['description'].apply(lambda x: word_tokenize(x))
content_series
```

```python

# content_series.loc[:,'words'] = content_series['description'].apply(lambda x: x.split())
# convert to lower case - starting with case-insensitive! - revisit this if high false positives
# content_series
```

```python

import string
# & -in the company's name -> P&G
puctuations_to_remove='!"#$%\'()*+,./:;<=>?@[\\]^_`{|}~' # needed puctuations
table = str.maketrans('', '', puctuations_to_remove)
def remove_puctuations(x):
    stripped = [w.translate(table) for w in x]
    return stripped

def lower_list(x):
    return [word.lower() for word in x]

content_series.loc[:,'stripped_w'] = content_series['words'].apply(lambda x: remove_puctuations(x))
content_series.loc[:,'stripped_w_lower'] = content_series['stripped_w'].apply(lambda x: lower_list(x))

# stripped = [w.translate(table) for w in tokens]
```
